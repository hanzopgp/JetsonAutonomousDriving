```python
import imageio
from io import BytesIO
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('data/test.mkv', 'rb') as file: 
    content = file.read()
```

```python
vid = imageio.get_reader(BytesIO(content),  'ffmpeg')
images = []
for num, image in enumerate(vid.iter_data()): 
    if num % 1000:
        images.append(image)
```

```python
print(len(images))
```
